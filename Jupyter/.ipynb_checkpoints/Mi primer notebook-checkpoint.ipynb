```python
 Si se sabe que cierta secuencia de instrucciones puede generar un problema, es posible añadir sentencias para que se ejecuten en caso de error, por ejemplo try y except
```

```python
grados = input ('¿Cuantos grados hay?')
try: 
    faren = float (grados)
    cel = (faren-32)* 5/9
    print (faren, 'grados farenheit son', cel, 'grados celsius' )

except:
    print ( 'Porfavor introduce un número')
```

```python
def mi_funcion(numero,exponente,cte):
    import math
    exp = (numero**exponente)-math.sin(cte)
    return(exp)
```

```python
mi_funcion(3,2,90)
```

```python
# Bucles tipo while
```

```python
n = 3
while n>0:
    print (n)
    n = n-1
print ("Accion")
```

```python
# Break
Permite salir de un bucle cuando se dese 
# Continue
Salta de una iteracion a otra 
```

```python
n = 3
while n>0:
    print (n)
    n = n-1
    if n == 2: break
print ("Accion")
```
